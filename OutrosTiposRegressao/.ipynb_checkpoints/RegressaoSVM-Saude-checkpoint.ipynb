# Regressão com SVM na base de dados Saúde

```python
# Numérico
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# Algoritmos de Aprendizagem
from sklearn.model_selection import train_test_split

# Support Vector Machines
from sklearn.svm import SVR
```

A classe `SVR`é usada tanto para problemas de **classificação** quanto problemas de **regressão**.

```python
# Carregar o dataframe da base
db_saude = pd.read_csv("plano_saude2.csv")

X_plano_saude = db_saude.iloc[:,0:1].values
y_plano_saude = db_saude.iloc[:,1].values
```

```python
X_plano_saude, y_plano_saude
```

Faremos o uso de três *kernels* diferentes:
1. Linear --- `svr_linear.fit(X_plano, y_plano)`
2. Polinomial
3. Radial Basis Function RBF

```python
### Instanciando os algoritmos
svr_linear = SVR(kernel='linear')
svr_poly = SVR(kernel='poly', degree=3)
svr_rbf = SVR(kernel='rbf')

### Treinamento dos Algoritmos
svr_linear.fit(X_plano_saude, y_plano_saude)
svr_poly.fit(X_plano_saude, y_plano_saude)
svr_rbf.fit(X_plano_saude, y_plano_saude)

### Armazenar as previsoes
pred_linear = svr_linear.predict(X_plano_saude)
pred_poly = svr_poly.predict(X_plano_saude)
pred_rbf = svr_rbf.predict(X_plano_saude)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, nrows=1, figsize=(12,5))

# Dados da base
ax1.scatter(x=X_plano_saude, y=y_plano_saude, c='red', s=5)
ax2.scatter(x=X_plano_saude, y=y_plano_saude, c='red', s=5)
ax3.scatter(x=X_plano_saude, y=y_plano_saude, c='red', s=5)

# Títulos
ax1.set_title('Kernel Linear')
ax2.set_title('Kernel Polinomial')
ax3.set_title('Kernel RBF')

# Plot dos ajustes
ax1.plot(X_plano_saude.ravel(), pred_linear)
ax2.plot(X_plano_saude.ravel(), pred_poly)
ax3.plot(X_plano_saude.ravel(), pred_rbf)
plt.tight_layout()
```

Os dados da nossa base **NÃO** estão **normalizados** (estão em escalas diferentes) e o **SVM** é sensível a isso, o que explica o resultado estranho do Kernel RBF.

### Normalização dos dados

```python
from sklearn.preprocessing import StandardScaler
```

```python
### Instanciar o 'Scalers' para X e y da base de dados
X_scaler, y_scaler = StandardScaler(), StandardScaler()
```

```python
# base original
X_plano_saude, y_plano_saude
```

```python
### Nova base aplicando o operador de escala
X_plano_new = X_scaler.fit_transform(X_plano_saude)
y_plano_new = y_scaler.fit_transform(y_plano_saude.reshape(-1,1))

X_plano_new.shape, y_plano_new.shape
```

Aqui $y$ foi passado como uma matriz no `fit_transform` mas nos algoritmos de previsão deve ser um vetor.

```python
X_plano_new, y_plano_new
```

```python
### Treinamento com a nova base -- y agora como vetor
svr_rbf.fit(X_plano_new, y_plano_new.ravel())

# Armazenar as novas previsoes
pred_rbf_new = svr_rbf.predict(X_plano_new)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, nrows=1, figsize=(12,5))

# Dados da base
ax1.scatter(x=X_plano_saude, y=y_plano_saude, c='red', s=5)
ax2.scatter(x=X_plano_saude, y=y_plano_saude, c='red', s=5)
ax3.scatter(x=X_plano_new, y=y_plano_new, c='red', s=5) ## Nova escala

# Títulos
ax1.set_title('Kernel Linear')
ax2.set_title('Kernel Polinomial')
ax3.set_title('Kernel RBF')

# Plot dos ajustes
ax1.plot(X_plano_saude.ravel(), pred_linear)
ax2.plot(X_plano_saude.ravel(), pred_poly)
ax3.plot(X_plano_new.ravel(), pred_rbf_new) ## Nova escala
plt.tight_layout()
```

**Previsão** deve ser feita levando em consideração a nova escala

Signature: y_scaler.inverse_transform(X, copy=None)  
Docstring:
Scale back the data to the original representation.


```python
svr_rbf.predict(
    [[40]]
)
```

```python
novo = [[40]]
```

```python
novo = X_scaler.transform(novo)
```

```python
novo
```

```python
# nova previsao com o registro escalonado
svr_rbf.predict(novo)
```

```python
# Voltando a escala original (transformação inversa) -- necessário formato de 
y_scaler.inverse_transform(svr_rbf.predict(novo).reshape(-1,1))
```
