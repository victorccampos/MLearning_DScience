```python
import os 
import pickle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open("census.pkl", 'rb') as file:    
    X_treinamento, y_treinamento, X_teste, y_teste = pickle.load(file)
```

5 vizinhos : 86%  
10 vinhos : 89%

```python
knn_census = KNeighborsClassifier(n_neighbors=10, metric='minkowski', p=2)
```

```python
# treinamento
knn_census.fit(X_treinamento, y_treinamento)
```

```python
#previsoes
y_pred = knn_census.predict(X_teste)
```

```python
# Precisão
sum(y_pred == y_teste) / y_teste.size
```

```python
cm = ConfusionMatrix(knn_census)
cm.fit(X=X_treinamento, y=y_treinamento)
cm.score(X_teste, y_teste)
```

```python
print(classification_report(y_true=y_teste, y_pred=y_pred))
```

O recall da classe >50K tá ruim (50%). Consegue identificar apenas 51% dos registros e quando identifica, a precisão é de 71%
