```python
import pandas as pd
import numpy as np

from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
```

```python
# Carregar o dataframe da base
db_saude = pd.read_csv("plano_saude2.csv")

X_plano_saude = db_saude.iloc[:,0:1].values
y_plano_saude = db_saude.iloc[:,1].values
```

Escalonamento das variáveis

```python
### Instanciar o 'Scalers' para X e y da base de dados
X_scaler, y_scaler = StandardScaler(), StandardScaler()

# Nova base aplicando o operador de escala
X_plano_new = X_scaler.fit_transform(X_plano_saude)
y_plano_new = y_scaler.fit_transform(y_plano_saude.reshape(-1,1))

X_plano_new.shape, y_plano_new.shape
```

```python
Neural_net_regressor = MLPRegressor(max_iter=1_000)
Neural_net_regressor.fit(X_plano_new, y_plano_new.ravel())
```

```python
# Avaliando na própria base de dados
Neural_net_regressor.score(X_plano_new, y_plano_new)
```

Se adapta _muito bem_ aos dados.

```python
y_previsto = Neural_net_regressor.predict(X_plano_new)
```

```python
### Plots

plt.scatter(X_plano_new, y_plano_new, c='red')
plt.plot(X_plano_new, y_previsto)
```
