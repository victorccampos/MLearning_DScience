```python
# Numérico
import pandas as pd
import numpy as np

# Gráficos
import matplotlib.pyplot as plt
import plotly.express as px 

# Algoritmos de Aprendizagem
from sklearn.model_selection import train_test_split

# Support Vector Machines
from sklearn.svm import SVR

# Metricas 
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
db_casas = pd.read_csv('house_prices.csv')

# predictor values in matrix shape
X_casas = db_casas.iloc[:,3:19].values

# prices
y_casas = db_casas.iloc[:,2].values

## Divisao da base
X_casas_treinamento, X_casas_test, y_casas_treinamento, y_casas_test = train_test_split(X_casas, y_casas, test_size=0.3, random_state=0)
X_casas_treinamento.shape, X_casas_test.shape
```

Para trabalhar com o Algoritmo de Support Vector Machine, precisamos fazer o escalonamento dos dados utilizando a classe `StandardScaler` utilizando o módulo de pré-processamento `sklearn.preprocessing`.

```python
from sklearn.preprocessing import StandardScaler
```

```python
### Scailing da base de treinamento

# Instanciando os Algoritmos para x e y
scaler_x: StandardScaler = StandardScaler()
scaler_y: StandardScaler = StandardScaler()

# Base de Treinamento : .fit_transform()
X_train_scaled = scaler_x.fit_transform(X_casas_treinamento)
y_train_scaled = scaler_y.fit_transform(y_casas_treinamento.reshape(-1,1))

# Base de Teste -- .transform() 
X_test_scaled = scaler_x.transform(X_casas_test)
y_test_scaled = scaler_y.transform(y_casas_test.reshape(-1,1)) 
```

```python
X_test_scaled # 16 colunas
```

```python
y_test_scaled # 1 coluna
```

```python
svm_regressor = SVR(kernel='rbf')
# Treinamento
svm_regressor.fit(X_train_scaled, y_train_scaled.ravel())
```

```python
score_svr_training = svm_regressor.score(X_train_scaled, y_train_scaled)
score_svr_testing = svm_regressor.score(X_test_scaled, y_test_scaled)
```

```python
score_svr_training
```

```python
score_svr_testing
```

score_svr_testing = 73% é melhor que regressão simples, múltipla, perde pra polinomial, resultado semelhante a árvore de decisão, perde pro RDF (88%)

```python
### Previsoes

y_pred = svm_regressor.predict(X_test_scaled)
```

```python
# Previsoes na escala normal
previsoes = scaler_y.inverse_transform(y_pred.reshape(-1,1)) 
```
