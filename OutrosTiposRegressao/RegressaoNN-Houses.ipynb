```python
# Numérico
import pandas as pd
import numpy as np

# Gráficos
import matplotlib.pyplot as plt
import plotly.express as px 

# Algoritmos de Aprendizagem
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Support Vector Machines
from sklearn.neural_network import MLPRegressor

# Metricas 
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
db_casas = pd.read_csv('house_prices.csv')

# predictor values in matrix shape
X_casas = db_casas.iloc[:,3:19].values

# prices
y_casas = db_casas.iloc[:,2].values

## Divisao da base
X_casas_treinamento, X_casas_test, y_casas_treinamento, y_casas_test = train_test_split(X_casas, y_casas, test_size=0.3, random_state=0)
X_casas_treinamento.shape, X_casas_test.shape
```

```python
### Scailing da base de treinamento

# Instanciando os Algoritmos para x e y
scaler_x: StandardScaler = StandardScaler()
scaler_y: StandardScaler = StandardScaler()

# Base de Treinamento : .fit_transform()
X_train_scaled = scaler_x.fit_transform(X_casas_treinamento)
y_train_scaled = scaler_y.fit_transform(y_casas_treinamento.reshape(-1,1))

# Base de Teste -- .transform() 
X_test_scaled = scaler_x.transform(X_casas_test)
y_test_scaled = scaler_y.transform(y_casas_test.reshape(-1,1)) 
```

Duas camadas ocultas e cada uma com 9 Neurônios (16+1/2).

```python
### Algoritmo NN com a base escalonada
# 16 -> 9 -> 9 -> 1
NN_regressor = MLPRegressor(max_iter=1_000, hidden_layer_sizes=(9,9))
NN_regressor.fit(X_train_scaled, y_train_scaled)
```

```python
score_NN_training = NN_regressor.score(X_train_scaled, y_train_scaled)
score_NN_testing = NN_regressor.score(X_test_scaled, y_test_scaled)
```

```python
score_NN_training, score_NN_testing
```

```python
# Previsoes na escala normal
previsoes = NN_regressor.predict(X_test_scaled) 
previsoes = scaler_y.inverse_transform(previsoes.reshape(-1,1))
```

```python
previsoes
```

```python
y_casas_test
```

```python
mean_absolute_error(y_true=y_casas_test, y_pred = previsoes)
```
