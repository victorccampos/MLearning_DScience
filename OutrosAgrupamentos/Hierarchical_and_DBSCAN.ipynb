```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

# Algoritmo Hierarquico
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Density Based Spatial Clustering of Applications with Noise
from sklearn.cluster import DBSCAN
```

Usaremos duas bases de dados:
1. **Base Salário** - (idade, salario)
2. **Base Cartão de Crédito** - "credit_card_clients.csv"

1 - Base Salário

```python
x_idades = [20,  27,  21,  37,  46, 53, 55,  47,  52,  32,  39,  41,  39,  48,  48]
y_salarios = [1000,1200,2900,1850,900,950,2000,2100,3000,5900,4100,5100,7000,5000,6500]  

base_salario = np.empty(shape=(len(x_idades),2),dtype=np.int32)

for k, (idade, salario) in enumerate(zip(x_idades, y_salarios)):
    base_salario[k] = [idade, salario]
base_salario

# Padronização dos Dados
scaler_salario = StandardScaler()
base_salario = scaler_salario.fit_transform(base_salario)
```

```python
base_salario
```

```python
plt.title("Base Salário")
plt.xlabel("Idades")
plt.ylabel("Salários")
plt.scatter(x=base_salario[:,0],  y=base_salario[:,1])
```

Utilizaremos método `linkage` do módulo `scipy.cluster.hierarchy` para fazer o agrupamento hierárquico com o Ward variance minimization algorithm.

```python
linkage_matrix = linkage(base_salario, method='ward')
dendograma = dendrogram(linkage_matrix)
plt.title("Dendrograma")
plt.xlabel("Pessoas")
plt.ylabel("Distância");
plt.axhline(y=3.5,color='k')
```

O critério de escolha do número de cluster é pegar a maior 'barra vertical' onde uma 'barra horizontal', aqui simbolizada pela linha preta, não cruze uma linha vertical. Dessa forma o número de clusters deve ser igual a 3.

```python
HC_salario = AgglomerativeClustering(n_clusters=3, linkage='ward', metric='euclidean')
labels_salario = HC_salario.fit_predict(base_salario)
labels_salario
```

```python
plt.title("Agrupamento Hierárquico")
plt.xlabel("Idades")
plt.ylabel("Salários")
## For a sequence of values to be color-mapped, use the 'c' argument instead.
plt.scatter(x=base_salario[:,0],  y=base_salario[:,1], c=labels_salario, cmap='viridis')
plt.colorbar()
plt.tight_layout()
```

`min_samples`:  
The number of samples (or total weight) in a neighborhood for a point to  be considered as a core point. This includes the point itself  .

If   `min_samples` is set to a higher value, DBSCAN will find denser cluster    whereas if it is set to a lower value, the found clusters will be m     sparse.

```python
dbscan_salario = DBSCAN(eps=0.5, min_samples=2)
dbscan_salario.fit(base_salario)
```

```python
dbscan_salario.labels_
```

O labels com -1 são considerados ruídos. O conjunto de parâmetros `num_samples=2` e $\varepsilon=0.5$ não foram boas escolhas para essa base de dados, o algoritmo não conseguiu atribuir grupos!

```python
dbscan_salario = DBSCAN(eps=0.95, min_samples=2)
dbscan_salario.fit(base_salario)
```

```python
dbscan_salario.labels_
```

Agora temos um agrupamento interessante.

```python
plt.title("Agrupamento DBSCAN")
plt.xlabel("Idades")
plt.ylabel("Salários")

plt.scatter(x=base_salario[:,0],  y=base_salario[:,1], c=dbscan_salario.labels_,cmap='viridis')
plt.colorbar()
plt.tight_layout()
```

2 - Base Crédito 

```python
base_cartao = pd.read_csv('credit_card_clients.csv', header=1)  # Header = 1 - Indica a presença de cabeçalho

### Criacao de Novo Atributo na Base de Dados
base_cartao['BILL_TOTAL'] = base_cartao['BILL_AMT1'] + base_cartao['BILL_AMT2'] + base_cartao['BILL_AMT3'] \
+ base_cartao['BILL_AMT4'] + base_cartao['BILL_AMT5'] + base_cartao['BILL_AMT6']

# Padronização dos Dados
X_cartao = base_cartao.iloc[:,[1, 25]].values
scaler_cartao = StandardScaler()
X_cartao = scaler_cartao.fit_transform(X_cartao)
```

```python
X_cartao
```

```python
linkage_cartao = linkage(X_cartao, method='ward')
dendrograma = dendrogram(Z=linkage_cartao)
```

O número ideal de clusters para essa base então é **3**

```python
### instanciando o algoritmo 
%timeit
HC_cartao = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
labels_cartao = HC_cartao.fit_predict(X_cartao)
```

```python
labels_cartao
```

```python
plt.figure(figsize=(12,5))
plt.title("Agrupamento Hierárquico - Base Cartão de Crédito")
plt.xlabel("Limite do Cartão")
plt.ylabel("Gastos - Bill Amount")

plt.scatter(x=X_cartao[:,0],  y=X_cartao[:,1], c=labels_cartao, cmap='inferno')
plt.colorbar()
plt.tight_layout()
```

```python
dbscan_cartao = DBSCAN(eps=0.5, min_samples=5)
labels_cartao_dbscan = dbscan_cartao.fit_predict(X_cartao)
```

Vamos contar quantos elementos há em cada grupo com `np.unique`

```python
np.unique(labels_cartao_dbscan, return_counts=True)
```

21 registros não foram classificados. É como se o algoritmo tivesse encontrado um grupo não classificado (21 registros) e outro grupo de 29979 como grupo 0.

```python
dbscan_cartao = DBSCAN(eps=0.37, min_samples=5)
labels_cartao_dbscan = dbscan_cartao.fit_predict(X_cartao)
np.unique(labels_cartao_dbscan, return_counts=True)
```

Esse foi um dos melhores resultados

```python
plt.figure(figsize=(12,5))
plt.title("Agrupamento DBSCAN - Base Cartão de Crédito ")
plt.xlabel("Limite do Cartão")
plt.ylabel("Gastos - Bill Amount")

plt.scatter(x=X_cartao[:,0],  y=X_cartao[:,1], c=labels_cartao_dbscan, cmap='inferno')
plt.colorbar()
plt.tight_layout()
```

De todos os algoritmos, o DBSCAN é o que mais se "testa" parâmetros que ajustem-se a uma base. Obs: até agora.
