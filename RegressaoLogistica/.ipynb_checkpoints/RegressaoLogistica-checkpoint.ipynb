```python
import os 
import pickle
import numpy as np

from sklearn.linear_model import LogisticRegression
from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import classification_report
```

#### Base Crédito

```python
with open(file='risco_credito.pkl', mode='rb') as file:
    X_risco_credito, y_risco_credito = pickle.load(file)
```

Excluir os registros "moderados

```python
y_risco_credito
# Indices = 2, 7, 11
```

```python
# Removendo os indices 2, 7, 11. axis = 0 (coluna)
X_risco_credito = np.delete(arr=X_risco_credito, obj=[2, 7, 11], axis = 0)
y_risco_credito = np.delete(arr=y_risco_credito, obj=[2, 7, 11], axis = 0)
```

```python
# Treinamento do Algoritmo
LR_risco_credit = LogisticRegression(random_state = 1)
LR_risco_credit.fit(X_risco_credito, y_risco_credito)
```

```python
# Parâmetro b0 : intercept
print(f'b0 = {LR_risco_credit.intercept_}')
# Parametros b1 = Historia, b2 = divida, b3 = garantias, b4 = renda anual.
print(f'b1, b2, b3, b4 = {LR_risco_credit.coef_}')
```

```python
# Testando com registro "artificiais"

y_pred_risco = LR_risco_credit.predict([
    [0,0,1,2],
    [2,0,0,0]
])
y_pred_risco
```
