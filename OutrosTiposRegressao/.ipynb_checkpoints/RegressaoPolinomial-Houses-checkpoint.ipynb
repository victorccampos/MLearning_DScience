```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px # gráficos dinâmicos

# Algoritmos de Aprendizagem
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Metricas 
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Graphic objects do Plotly
#import plotly.graph_objects as go
```

O método fit apenas ajusta os parâmetros do scaler, mas não retorna nada. fit_transform é utilizado quando queremos ajustar esses parâmetros mas também retornar os próprios valores escalonados; é equivalente a utilizar fit depois transform. transform é utilizado quando o scaler já está ajustado, e apenas queremos que os valores sejam transformados.

```python
db_casas = pd.read_csv('house_prices.csv')
db_casas
```

“flatten” cria uma cópia do array e retorna um array unidimensional, enquanto “ravel” retorna uma visão do array.

```python
# square foot living values in matrix shape
X_casas = db_casas.iloc[:,3:19].values
# predictors: price
y_casas = db_casas.iloc[:,2].values
```

```python
X_casas_treinamento, X_casas_teste, y_casas_treinamento, y_casas_test = train_test_split(X_casas, y_casas, test_size=0.3, random_state=0)
```

```python
X_casas_treinamento.shape, X_casas_teste.shape
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
X_casas_treinamento_poly = poly.fit_transform(X_casas_treinamento)
X_casas_teste_poly = poly.transform(X_casas_teste)
```

```python
X_casas_treinamento_poly.shape, X_casas_teste_poly.shape
```

```python
regressor_casas_poly = LinearRegression()
regressor_casas_poly.fit(X_casas_treinamento_poly, y_casas_treinamento)
```

```python
regressor_casas_poly.score(X_casas_treinamento_poly, y_casas_treinamento)
```

```python
regressor_casas_poly.score(X_casas_teste_poly, y_casas_test)
```

```python
previsoes = regressor_casas_poly.predict(X_casas_teste_poly)
previsoes
```

```python
y_casas_test
```

```python
## Na video-aula 165.884,00
mean_absolute_error(y_casas_test, previsoes)  
```
