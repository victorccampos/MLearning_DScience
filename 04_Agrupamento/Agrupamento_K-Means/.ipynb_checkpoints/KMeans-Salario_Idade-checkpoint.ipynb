```python
import plotly.express as px
import plotly.graph_objects as go

import numpy as np
import matplotlib.pyplot as plt

# Pré-Processamento
from sklearn.preprocessing import StandardScaler
# K-Means
from sklearn.cluster import KMeans
```

# Base de idade e salário

```python
# X - idade y- salarios
x_idades = [20,  27,  21,  37,  46, 53, 55,  47,  52,  32,  39,  41,  39,  48,  48]
y_salarios = [1000,1200,2900,1850,900,950,2000,2100,3000,5900,4100,5100,7000,5000,6500]  
```

```python
plt.scatter(x=x_idades, y=y_salarios);
```

```python
base_salario = np.empty(shape=(len(x_idades),2),dtype=np.int32)
```

```python
for k, (idade, salario) in enumerate(zip(x_idades, y_salarios)):
    base_salario[k] = [idade, salario]
base_salario
```

Padronização dos dados

```python
scaler_salario = StandardScaler()
base_salario = scaler_salario.fit_transform(base_salario)
```

```python
# Base Padronizada
base_salario
```

```python
### Instanciando o Algoritmo
k_means = KMeans(n_clusters=3)
k_means.fit(base_salario)
```

```python
# Centroides - Média dos registros que compõe um grupo
k_means.cluster_centers_
```

```python
scaler_salario.inverse_transform(k_means.cluster_centers_)
```

```python
k_means.labels_
```

```python

plt.title("Agrupamento K-means")
plt.xlabel("Idades")
plt.ylabel("Salários")

plt.scatter(x=base_salario[:,0],  y=base_salario[:,1], c=k_means.labels_,cmap='viridis')
plt.scatter(x=k_means.cluster_centers_[:,0], y= k_means.cluster_centers_[:,1], marker='X', s=100, label='Centroide')
plt.colorbar()
plt.legend()
plt.tight_layout()
```
