```python
# Plotting and numerical
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi']=300 # highres display

from yellowbrick.classifier import ConfusionMatrix
# Files
import os 
import pickle

# sklearn imports 
from sklearn import tree
from sklearn.metrics import classification_report, accuracy_score
```

- 1 : Carregar o arquivo do pickle
- 2 : Criar a árvore
- 3: Treinar
- 4: Fazer as previsões 
- 5: Avaliar as previsões
    - `accuracy_score`
    - `matriz de confusão` 
    - `visualizar árvore`

```python
with open(file="census.pkl", mode='rb') as file:
    X_treinamento, y_treinamento, X_teste, Y_teste = pickle.load(file)
print("\t Divisão da Base")
print(f"Treinamento : {X_treinamento.shape, y_treinamento.shape}")
print(f"Teste : {X_teste.shape, Y_teste.shape}")
```

```python
# criar arvore
census_tree = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
# treinar a arvore
census_tree.fit(X_treinamento, y_treinamento)
# Previsao - receber o teste
y_pred = census_tree.predict(X_teste)
```

```python
# Avaliar as previsoes
print(f'Precisão : {accuracy_score(y_true=Y_teste, y_pred=y_pred)}')

cm = ConfusionMatrix(estimator=census_tree)
cm.fit(X_treinamento, y_treinamento)
cm.score(X_teste, Y_teste)
```

Da classe $\le 50K$, 3227 foram classificados corretamente e 466 foram classificados erroneamente $>50K$.  
Da classe $>50K$, 732 foram classificados corretamente e 460 foram classificados erroneamente. Para essa classe, o resultado não foi muito bom.

```python
print(classification_report(y_true=Y_teste, y_pred=y_pred))
```
