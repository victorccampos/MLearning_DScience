```python
import pickle 
import matplotlib.pyplot as plt
# Multi Layer Perceptron Classifier
from sklearn.neural_network import MLPClassifier

# Métricas de desempenho
from sklearn.metrics import accuracy_score, classification_report  # Métricas de  Classificacao
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay # Métricas de Classificacao
```

#### Base de dados Crédito

```python
with open("credit.pkl", 'rb') as file:
    X_train, y_train, X_test, y_test = pickle.load(file)
```

**Estrutura da rede**: 3 -> 100 -> 100 -> 1  

```python
### Instanciar o algoritmo 
# verbose - mostrar o loss a cada época
# tol - de época em época o erro tem que diminuir em 'tol'
multi_layer_percep = MLPClassifier(max_iter = 1_500, verbose=True, tol=0.00000100, solver='adam', hidden_layer_sizes=(100,100), activation='relu')
multi_layer_percep.fit(X_train, y_train)
```

*ConvergenceWarning*: indica que o algoritmo ainda pode diminuir o valor do erro e estamos executando com poucas épocas. 

```python
y_pred = multi_layer_percep.predict(X_test)
```

```python
matriz_confusao = confusion_matrix(y_true = y_test,
                                  y_pred = y_pred
                                  , labels= None)

figure = plt.figure(figsize=(15,5))
disp = ConfusionMatrixDisplay(confusion_matrix=matriz_confusao, display_labels=None)
disp.plot(values_format='d')
plt.show()
```

```python
# Metricas de precisão, revocação, f1-score e acurácia.
print(classification_report(y_test, y_pred))
```
