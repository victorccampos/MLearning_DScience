```python
import os 
import pickle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from yellowbrick.classifier import ConfusionMatrix
```

Os dados já foram pré-processados e fez-se a **padronização**

```python
with open("credit.pkl", 'rb') as file:    
    X_treinamento, y_treinamento, X_teste, y_teste = pickle.load(file)
```

```python
knn_credit = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
```

A métrica de minkowski com $p=2$ equivale a distância euclidiana

```python
# treinamento
knn_credit.fit(X_treinamento, y_treinamento)
```

```python
#previsoes
y_pred = knn_credit.predict(X_teste)
```

```python
# Precisão
sum(y_pred == y_teste) / y_teste.size
```

```python
cm = ConfusionMatrix(knn_credit)
cm.fit(X=X_treinamento, y=y_treinamento)
cm.score(X_teste, y_teste)
```

```python
print(classification_report(y_true=y_teste, y_pred=y_pred))
```
