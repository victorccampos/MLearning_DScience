```python
import plotly.express as px
import plotly.graph_objects as go

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Pré-Processamento
from sklearn.preprocessing import StandardScaler
# K-Means
from sklearn.cluster import KMeans
```

**Documentação do Agrupamento KMeans** : [https://scikit-learn.org/stable/modules/clustering.html#k-means]

Uma base de pagamento de cartão de crédito de 30 000 clientes em um período de 6 meses.  
1. `BILL_AMT%` - o valor total fatura do mês
2. `PAY_AMT%`  - quanto o cliente *pagou*
3. `default payment next month` - **1**: não pagou **0** pagou

```python
# Header = 1 --- indica a presença de cabeçalho
base_cartao = pd.read_csv('credit_card_clients.csv', header=1)
base_cartao
```

```python
### Criacao de Novo atributo na Base de Dados
base_cartao['BILL_TOTAL'] = base_cartao['BILL_AMT1'] + base_cartao['BILL_AMT2'] + base_cartao['BILL_AMT3'] \
+ base_cartao['BILL_AMT4'] + base_cartao['BILL_AMT5'] + base_cartao['BILL_AMT6']
```

```python
base_cartao
```

Criaremos uma variável chamada `X_cartao` que irá ter os registros do limite do cartão (coluna 1) e a fatura total (coluna 2).

```python
X_cartao = base_cartao.iloc[:,[1, 25]].values
X_cartao
```

**Padronização dos Dados**

```python
scaler_cartao = StandardScaler()

X_cartao = scaler_cartao.fit_transform(X_cartao)
```

```python
WCSS = []

## Guessing k's
for i in range(1,11):
    kmeans_cartao = KMeans(n_clusters=i, random_state=0)
    kmeans_cartao.fit(X_cartao)
    # 
    WCSS.append(kmeans_cartao.inertia_)
WCSS
```

```python
plt.title('Optimal $k$ - "Elbow Method" ')
plt.ylabel("Inertia")
plt.xlabel(r"$k$ " + "value")
plt.scatter(x=range(1,11), y=WCSS);
```

Podemos escolher o agrupamento ou com 4 ou com 5.

### Algoritmo Final

```python
kmeans_card = KMeans(n_clusters=4, random_state=0)
rotulos = kmeans_card.fit_predict(X_cartao)
```

```python
rotulos
```

```python
plt.figure(figsize=(12,5))
plt.title("Agrupamento K-means")
plt.xlabel("Limite do Cartão")
plt.ylabel("Gastos - Bill Amount")

plt.scatter(x=X_cartao[:,0],  y=X_cartao[:,1], c=rotulos, cmap='inferno')
#plt.scatter(x=k_means.cluster_centers_[:,0], y= k_means.cluster_centers_[:,1], marker='X', s=100, label='Centroide')
plt.colorbar()
plt.tight_layout()
```

Separação dos Clientes via GRUPO (0, 1, 2, 3) com  $k=4$.  
1. Criação de uma base única: `np.column_stack(db_original, rotulos-k-means)`
2. Ordenação de toda a base, com base no critério do GRUPO via `argsort`

```python
LISTA_CLIENTES = np.column_stack((base_cartao, rotulos))
LISTA_CLIENTES
```

numpy `argsort()`: Returns the indices that would sort this array.  
Exemplo: [2,4,1,3] ao ser ordenado "viraria" [1,2,3,4]. Então, o retorno de argsort, com base em [2,4,1,3] é `[1,3,0,2]`.  
Lê-se:  
**2** deve estar na posiçao 1  
**4** na posição 3  
**1** na posição 0  
**3** na posição 2.


```python
LISTA_CLIENTES[:,26].argsort()
```

```python
## Sorted by index
LISTA_CLIENTES = LISTA_CLIENTES[
    LISTA_CLIENTES[:,26].argsort()
]
```

```python
LISTA_CLIENTES
```

## Agrupamento com mais parâmetros

```python
base_cartao.head(1)
```

```python
print(base_cartao.columns)
X_new = base_cartao.iloc[:,[1,2,3,4,5,25]].values
X_new
```

```python
## Padronização dos dados
scaler_x_new = StandardScaler()
X_new = scaler_x_new.fit_transform(X_new)
X_new
```

```python
WCSS = []

## Guessing k's
for i in range(1,11):
    kmeans_new = KMeans(n_clusters=i, random_state=0)
    kmeans_new.fit(X_new)
    # 
    WCSS.append(kmeans_new.inertia_)
WCSS
```

```python
plt.title('Optimal $k$ - "Elbow Method" ')
plt.ylabel("Inertia")
plt.xlabel(r"$k$ " + "value")
plt.grid()
plt.scatter(x=range(1,11), y=WCSS);
```

```python
kmeans_2 = KMeans(n_clusters=2, random_state=0)
labels_k2= kmeans_2.fit_predict(X_new)
```

Como temos 6 atributos, fazemos **Principal Component Analysis (PCA)** para a redução de dimensionalidade. De *6* atributos sairemos para apenas *2* atributos.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_principal = pca.fit_transform(X_new)
X_principal.shape
```

```python
X_principal
```

```python
plt.figure(figsize=(12,5))
plt.title("Agrupamento K-means")

plt.scatter(x=X_principal[:,0],  y=X_principal[:,1], c=labels_k2, cmap='inferno')
#plt.scatter(x=k_means.cluster_centers_[:,0], y= k_means.cluster_centers_[:,1], marker='X', s=100, label='Centroide')
plt.colorbar()
plt.tight_layout()
```
